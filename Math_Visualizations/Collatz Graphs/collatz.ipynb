```python
from matplotlib import pyplot as plt
from math import sqrt
from sympy import primefactors
import numpy as np
```

```python
def Collatz(n, count = 0):
    while n != 1:
        if n % 2 == 0:
            n /= 2
        else:
            n *= 3
            n += 1
        count += 1
    return count
    
```

```python
CollatzList = []
for n in range(1, 101):
    CollatzList.append(Collatz(n))
CollatzList = np.array(CollatzList)
Xticks = np.arange(1, 101)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(Xticks, CollatzList, c=CollatzList, cmap='Spectral')
plt.xlabel('Number')
plt.ylabel('Collatz Number')
plt.show()
```

```python
CollatzExpanded = np.array([Collatz(n) for n in range(1,100001)])
X_Expanded = np.arange(1,100001)
size = [2 for i in range(100000)]
plt.scatter(X_Expanded, CollatzExpanded, c=CollatzExpanded, cmap='Spectral', s=size)
plt.xlabel('Number')
plt.ylabel('Collatz Number')
plt.show()
```

```python
primefactorcount = np.array([len(primefactors(n)) for n in range(1,101)])

fig = plt.figure()

ax1 = fig.add_subplot(111)
ax1.scatter(Xticks, CollatzList, s=15, c='r', label='Collatz Number')
ax1.tick_params(axis='y', labelcolor='r')
ax1.set_xlabel('Number')
ax1.set_ylabel('Collatz Number')

ax2 = ax1.twinx()
ax2.scatter(Xticks, primefactorcount, s=15, c='b', label='Prime Factor Count')
#No need to set X label as its already been done
ax2.set_ylabel('Number of Prime Factors')
ax2.tick_params(axis='y', labelcolor='b')

plt.legend(loc='upper left')
plt.show() # This is a suuuuuper boring graph!!!
```

```python
primefactorcountexpanded = np.array([len(primefactors(n)) for n in range(1,100001)])
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(X_Expanded, CollatzExpanded, s=2, c='r', label="Collatz Number")
ax1.set_ylabel('Collatz Number')
ax1.set_xlabel('Number')
ax1.tick_params(axis='y', labelcolor='r')

ax2 = ax1.twinx()
ax2.scatter(X_Expanded, primefactorcountexpanded, s=2, c='b', label="Prime Factor Count")
ax2.set_ylabel("Number of Prime Factors")
ax2.tick_params(axis='y', labelcolor='b')

plt.legend(loc='upper left')
plt.show() # Another Boring Graph!!!!
```
